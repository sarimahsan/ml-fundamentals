```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constraints
def constraint1(x):
    return (50 - 5*x) / 10

def constraint2(x):
    return (14 - 7*x) / 2

def constraint3(x):
    return (3*x - 6) / 2

# Define the range of x values
x_values = np.linspace(0, 10, 400)

# Plot the constraints
plt.plot(x_values, constraint1(x_values), label=r'$5x_1 + 10x_2 \leq 50$')
plt.plot(x_values, constraint2(x_values), label=r'$7x_1 + 2x_2 \leq 14$')
plt.plot(x_values, constraint3(x_values), label=r'$3x_1 - 2x_2 \geq 6$')

# Plot non-negativity constraints
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Fill the feasible region
plt.fill_between(x_values, 0, constraint1(x_values), where=(constraint1(x_values) >= 0) & (x_values >= 0), color='gray', alpha=0.5)
plt.fill_between(x_values, 0, constraint2(x_values), where=(constraint2(x_values) >= 0) & (x_values >= 0), color='gray', alpha=0.5)
plt.fill_between(x_values, 0, constraint3(x_values), where=(constraint3(x_values) >= 0) & (x_values >= 0), color='gray', alpha=0.5)

# Annotate corner points
plt.scatter([0, 10, 0, 2], [5, 0, 7, 0], color='red')
plt.text(0, 5, 'A (0, 5)', ha='right')
plt.text(10, 0, 'B (10, 0)', ha='right')
plt.text(0, 7, 'C (0, 7)', ha='right')
plt.text(2, 0, 'D (2, 0)', ha='right')

# Set plot labels and legend
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Feasible Region')
plt.legend()

# Show plot
plt.grid(True)
plt.xlim(0, 12)
plt.ylim(0, 10)
plt.show()
```
